# Let's draw things on image

# Draw lines on the image 

Use ``line`` in opencv to draw line in one of your images.

It should be like this form ``cv2.line(image, start_point, end_point, color, line_width)``. You may check whatever doc you want.

color in opencv is in BGR format.
<br>
``(255, 255, 255) -> white``
<br>
``(0,0,0)         -> Black``
<br>
``(255,0,0) -> Blue``
<br>
``(0,255,0) -> Green``
<br>
``(0,0,255) -> Red``


```python
cv2.line(image , (0,0) , (511 , 773) , (255,0,0),5)

cv2.imshow("Lady_GaGa" , image)

cv2.imwrite("C:/Users/MOHAMMAD AMIN/Desktop/my-notebook/lagy-gaga2-line.jpg" , image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

# Draw rectangle on the image

Use ``rectangle`` in opencv to draw a rectangle in one of your images

```python
cv2.rectangle(image1,(173,111),(600,400),(0,255,0),3)

cv2.imshow("Lady_GaGa" , image1)

cv2.imwrite("C:/Users/MOHAMMAD AMIN/Desktop/my-notebook/lagy-gaga-rectangle.jpg" , image1)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

# Draw circle on the image
Use ``circle`` in opencv to draw a circle on your image.

```python
cv2.circle(image,(256,387), 100, (0,0,255), -1)

cv2.imshow("Lady_GaGa" , image)

cv2.imwrite("C:/Users/MOHAMMAD AMIN/Desktop/my-notebook/lagy-gaga2-line-circle.jpg" , image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

# Connect points with lines

Run cells below, and see what happens.

```python
pts = np.array([[10, 5], [20, 30], [70, 40], [90, 120], [50, 5]], np.int32)
```

$True$ means you want the last point to be connected to the first point too.

```python
cv2.polylines(image, [pts], True, (0, 255, 255), 5)

cv2.imshow("Lady_GaGa" , image)

cv2.imwrite("C:/Users/MOHAMMAD AMIN/Desktop/my-notebook/lagy-gaga2-line-circle-polylines.jpg" , image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
